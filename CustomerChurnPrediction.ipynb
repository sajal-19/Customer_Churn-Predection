## Data Cleaning

```python
telecom=telecom_data.copy()
```

```python
# Coverting TotalCharges to Numeric DataType
telecom.TotalCharges = pd.to_numeric(telecom.TotalCharges, errors='coerce')
telecom.isnull().sum()
```

11 Missing Values in TotalCharges

```python
#Show records having Missing Values
telecom.loc[telecom['TotalCharges'].isnull() == True]
```

#### Removing Missing Values

```python
telecom.dropna(how = 'any', inplace = True)
```

```python
telecom.loc[telecom['TotalCharges'].isnull() == True]
```

No Missing Values in the dataset

```python
#Max tenure in the datset
print(telecom['tenure'].max())
```

```python
# Grouping tenure in bins of 12 months
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]

telecom['tenure_group'] = pd.cut(telecom.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
#Groups of tenure
telecom['tenure_group'].value_counts()
```

```python
#Removing columns that are not required now
telecom.drop(columns= ['customerID','tenure'], axis=1, inplace=True)
```

## Data Exploration

Univariate Analysis using ploting

```python
#  Plot distibution of individual predictors by churn
for i, predictor in enumerate(telecom.drop(columns=['Churn', 'TotalCharges', 'MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data=telecom, x=predictor, hue='Churn')
```

```python
# Covert the target value 'Churn' in Binary Numeric Value yes(1),No(0)
telecom['Churn'] = np.where(telecom.Churn == 'Yes',1,0)
```

```python
# data
telecom.head()
```

```python
# Converting all categorical variables to dummy variables
telecom_dummies = pd.get_dummies(telecom)
telecom_dummies.head()
```

Showing Relationship Between MonthlyCharges and TotalCharges using Ploting

```python
sns.lmplot(data=telecom_dummies, x='MonthlyCharges', y='TotalCharges', fit_reg=False)
```

Result: TotalCharges increases as MonthyCharges increases

#### Churning by MonthlyCharges and TotalCharges

```python
# Churning By Monthly Charges
Mth = sns.kdeplot(telecom_dummies.MonthlyCharges[(telecom_dummies["Churn"] == 0) ],color="Red", shade=True)
Mth = sns.kdeplot(telecom_dummies.MonthlyCharges[(telecom_dummies["Churn"] == 1) ],ax =Mth, color="Blue", shade=True)
Mth.legend(["No Churn","Churn"],loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Monthly Charges')
Mth.set_title('Monthly charges by churn')
```

From Graph We can analyze churn is high when monthly charges are high

```python
# Churn By Total Charges
Tot = sns.kdeplot(telecom_dummies.TotalCharges[(telecom_dummies["Churn"] == 0) ],
                color="Red", shade = True)
Tot = sns.kdeplot(telecom_dummies.TotalCharges[(telecom_dummies["Churn"] == 1) ],
                ax =Tot, color="Blue", shade= True)
Tot.legend(["No Churn","Churn"],loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Total Charges')
Tot.set_title('Total charges by churn')
```

Insight: Higher Churn at Lower Total Charges

Result: Higher Monthly Charges,lower tenure,lower total charges resulting to high churn

#### Bivariate Analysis using Ploting

```python
new_df1_target0=telecom.loc[telecom["Churn"]==0]
new_df1_target1=telecom.loc[telecom["Churn"]==1]
```

```python
def uniplot(df,col,title,hue =None):
    
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width , 8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title)
    ax = sns.countplot(data = df, x= col, order=df[col].value_counts().index,hue = hue,palette='bright') 
        
    plt.show()
```

```python
uniplot(new_df1_target1,col='Partner',title='Distribution of Gender for Churned Customers',hue='gender')
```

```python
uniplot(new_df1_target0,col='Partner',title='Distribution of Gender for Non Churned Customers',hue='gender')
```
